### Limpieza de datos

```python
#  Cleaning HISTORICO_SUERTES.xlsx
import pandas as pd

# load raw data
df_hist = pd.read_excel('HISTORICO_SUERTES.xlsx')

# make a copy to clean
df_hist_clean = df_hist.copy()

# Eliminación o imputación de columnas con muchos nulos (>30%)
cols_to_drop_hist = df_hist_clean.columns[df_hist_clean.isnull().mean() > 0.3]
df_hist_clean.drop(columns=cols_to_drop_hist, inplace=True)


#  Clip outliers in TCH and %Sac.Caña to the [1st,99th] percentiles
for col in ['TCH', '%Sac.Caña']:
    low = df_hist_clean[col].quantile(0.01)
    high = df_hist_clean[col].quantile(0.99)
    df_hist_clean[col] = df_hist_clean[col].clip(lower=low, upper=high)

#Impute missing numeric values with the median
numeric_cols = df_hist_clean.select_dtypes(include='number').columns
for col in numeric_cols:
    median = df_hist_clean[col].median()
    df_hist_clean[col].fillna(median, inplace=True)

# Imputación simple para valores faltantes restantes
df_hist_clean.fillna(method='ffill', inplace=True)


# Export cleaned dataset

df_hist_clean.to_excel('HISTORICO_SUERTES_clean.xlsx', index=False)

```

```python
# Cleaning BD_IPSA_1940.xlsx
import pandas as pd

# load raw data
df_ipsa = pd.read_excel('BD_IPSA_1940.xlsx')

# make a copy to clean
df_ipsa_clean = df_ipsa.copy()

#  Clip outliers in TCH and sacarosa to the [1st,99th] percentiles
for col in ['TCH', 'sacarosa']:
    low = df_ipsa_clean[col].quantile(0.01)
    high = df_ipsa_clean[col].quantile(0.99)
    df_ipsa_clean[col] = df_ipsa_clean[col].clip(lower=low, upper=high)

#  Impute missing numeric values with the median
numeric_cols = df_ipsa_clean.select_dtypes(include='number').columns
for col in numeric_cols:
    median = df_ipsa_clean[col].median()
    df_ipsa_clean[col].fillna(median, inplace=True)

#  Export cleaned dataset
df_ipsa_clean.to_excel('BD_IPSA_1940_clean.xlsx', index=False)

```

```python

# Usando percentiles para categorizar TCH y Sacarosa
tch_quantiles = ipsa_df['TCH'].quantile([0.33, 0.66])
sac_quantiles = ipsa_df['sacarosa'].quantile([0.33, 0.66])

def clasificar(valor, q1, q2):
    if valor < q1:
        return 'Bajo'
    elif valor < q2:
        return 'Medio'
    else:
        return 'Alto'

ipsa_df['TCH_cat'] = ipsa_df['TCH'].apply(lambda x: clasificar(x, tch_quantiles[0.33], tch_quantiles[0.66]))
ipsa_df['sacarosa_cat'] = ipsa_df['sacarosa'].apply(lambda x: clasificar(x, sac_quantiles[0.33], sac_quantiles[0.66]))
```

```python
# Visualizamos la distribución de categorías
sns.countplot(data=ipsa_df, x='TCH_cat', palette='viridis')
plt.title('Categorías de TCH')
plt.show()

sns.countplot(data=ipsa_df, x='sacarosa_cat', palette='plasma')
plt.title('Categorías de Sacarosa')
plt.show()
```
