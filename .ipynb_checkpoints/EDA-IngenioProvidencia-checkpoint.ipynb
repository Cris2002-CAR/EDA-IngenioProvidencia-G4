```python

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Carga de los datasets
historico_df = pd.read_excel("HISTORICO_SUERTES.xlsx")
ipsa_df = pd.read_excel("BD_IPSA_1940.xlsx")
```

```python

# Mostrar las primeras filas
historico_df.head()
```

```python

# Información y nulos para HISTORICO_SUERTES
historico_df.info()
historico_df.isnull().sum()
```

```python

# Información y nulos para BD_IPSA_1940
ipsa_df.info()
ipsa_df.isnull().sum()
```

```python

# Distribución de TCH y % Sacarosa en HISTORICO_SUERTES
sns.histplot(historico_df['TCH'], kde=True)
plt.title("Distribución de TCH (HISTORICO_SUERTES)")
plt.show()

sns.histplot(historico_df['Ton.Azucar'], kde=True)
plt.title("Distribución de Toneladas de Azúcar (como proxy de sacarosa)")
plt.show()
```

```python

# Distribución de TCH y Sacarosa en BD_IPSA_1940
sns.histplot(ipsa_df['TCH'], kde=True)
plt.title("Distribución de TCH (BD_IPSA_1940)")
plt.show()

sns.histplot(ipsa_df['sacarosa'], kde=True)
plt.title("Distribución de Sacarosa (BD_IPSA_1940)")
plt.show()
```
